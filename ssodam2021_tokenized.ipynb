```python
import pandas as pd
from konlpy.tag import Okt
```

```python
import os
os.listdir()
```

```python
sd = pd.read_csv('./data/ssodam2021half+popularity.csv')
sd
```

```python
sd.isnull().sum()
```

```python
ssodam = sd.dropna(thresh = 7)
ssodam.reset_index(drop = True, inplace = True)
ssodam
```

```python
ssodam[ssodam['title'].isnull() == True]
```

```python
ssodam['text'] = ssodam['text'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣, \d, % ]","")
ssodam['title'] = ssodam['title'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣, \d, % ]","")
ssodam['comments'] = ssodam['comments'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣, \d, % ]","")
ssodam['comments_writer'] = ssodam['comments_writer'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣, \d, % ]","")
```

```python
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','을','안','한','로',',','아','다미','부터','이랑',
             '자','에','와','하다','ㅋ','명','것','곳','거','때','그','데','못','중','대한','인데','진짜','너무','에서','근데']
```

```python
okt = Okt()

def tokn(x):
    try:
        temp_y = okt.pos(x) # 토큰화
        temp_y = [word for word in temp_y if not word in stopwords] # 불용어 제거
        return temp_y
    except:
        print(x)
```

```python
ssodam['text_tokenized'] = ssodam['text'].apply(lambda x : tokn(x))
ssodam['title_tokenized'] = ssodam['title'].apply(lambda x : tokn(x))
ssodam['comments_tokenized'] = ssodam['comments'].apply(lambda x : tokn(x))
ssodam['comments_writer_tokenized'] = ssodam['comments_writer'].apply(lambda x : tokn(x))
```

```python
ssodam.to_csv('ssodam2021half_tokenized.csv', encoding = 'utf-8', index = False)
```
